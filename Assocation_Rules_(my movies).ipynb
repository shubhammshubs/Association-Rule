# 1. Import Libs 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings('ignore')
```

# 2. Import Data 

```python
movie = pd.read_csv('my_movies.csv')
movie
```

```python
movie.isna().sum()
```

```python
movie.dtypes
```

#### As unneccery data is resent in 1st v5 columns so we will not take those columns for processing 

```python
movie2=movie.iloc[:,5:]
movie2
```

```python
movie2.dtypes
```

# 3. Model Building 

## Apriorri Alogrithm 

### 1. Association rules with 10% Support and 70% confidenc 

#### 1.1 With 10% Support 

```python
frequent_itemsets=apriori(movie2,min_support=0.1,use_colnames=True)
frequent_itemsets
```

#### 1.2 with 70% confidence 

```python
association_1=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
association_1
```

```python
association_1[association_1.lift>1]
```

## visualization

```python
plt.scatter(association_1['support'],association_1['confidence'],c= association_1['lift'], cmap = 'twilight_shifted')
plt.colorbar()
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
ax1 = plt.figure(figsize=(10,5)).add_subplot(111 ,projection = '3d')
ax1.scatter(association_1['support'],association_1['confidence'],association_1['lift'])
ax1.set_xlabel("support")
ax1.set_ylabel("confidence")
ax1.set_zlabel("lift")
```
